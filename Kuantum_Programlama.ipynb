# KUANTUM PROGRAMLAMA

Kuantum programlama, kuantum bilgisayarlar gibi kuantum mekanik prensipleri kullanarak bilgi işleme ve hesaplama yapmayı amaçlayan bir programlama paradigmasıdır. Kuantum bilgisayarlar, klasik bilgisayarlardan farklı olarak kuantum bit (qubit) adı verilen kuantum mekanik özelliklere dayanan bilgi birimlerini kullanırlar.

Kuantum programlamada amaç, kuantum algoritmalarını ve kuantum devrelerini tasarlayarak belirli problemleri çözmektir. Kuantum bilgisayarlar, belirli problemleri daha etkili bir şekilde çözebilecekleri düşünülen bazı algoritmalar için potansiyel avantajlar sunabilirler. Örneğin, Shor'un algoritması sayesinde, kuantum bilgisayarlar klasik bilgisayarlarla çözülemeyen bazı sayı teorisi problemlerini daha hızlı çözebilirler.

Ancak, kuantum bilgisayarların güçlü ve geniş ölçekli kullanımı hala birçok teknik zorlukla karşı karşıyadır. Kuantum hallerinin korunması (kuantum coerence), kuantum gürültüsü ve qubit'lardaki hatalar gibi sorunlar, kuantum programlamanın gelişmesini sınırlayan zorluklardan sadece birkaçıdır. Bu nedenle, kuantum programlama şu anda genellikle sınırlı sayıda qubit içeren küçük ölçekli kuantum bilgisayarlar üzerinde odaklanmaktadır.


# **Bit** **ve** **Qubit** **Farkı**

Klasik bilgisayarlarda bilgi bitler aracılığı ile işlenir ve bitler 0 veya 1 değerlerini alır. Fakat kuantum bilgisayarlarda qubitler 0,1 değerlerini almasının yanı sıra 0 ve 1 değerini aynı anda alabilir. Buna *süperpozisyon ilkesi* denilir. Bu kuantum fenomeni sayesinde klasik bir bilgisayarın hesaplayabileceği bir sorunu kuantum bilgisayarlar daha hızlı hesaplayabilirler. Bu nasıl oluyor diye sorarsanız qubitler yine kuantum mekaniğinde yer alan bir diğer fenomen *dolanıklık* sayesinde bunu gerçekleştiriyorlar. Kuantum mekaniğinde eğer bir parçacığın o anki durumunu ölçmez iseniz hangi durumda olduğunu bilemezsiniz, sadece olasılığını hesaplarsınız. Dolanıklığı ise şöyle açıklayabiliriz; , iki ya da daha fazla parçacığın birbirleriyle öylesine ilişkili olması durumunu ifade eder ki, bir parçacığın durumu diğerini anında etkiler, mesafeye bağlı olmaksızın. Gerçek hayatta, iki dolanıklı parçacık arasında ölçüm yapıldığında, birinin durumu anında diğerini belirler. Bu, klasik fizik kurallarının ötesinde bir etkileşimdir ve Einstein bunu "uzak etkileme" olarak adlandırmıştır.

# **Bilmemiz** **Gerekenler**



1.   **Süperpozisyon**: Bir sistemin aynı anda hem 1 hem 0 olması durumuna denir.
2.  **Girişim**: Kuantum dalga fonksiyonlarının birleşmesi veya çakışması sonucunda oluşur. Kuantum durumlarını güçlendirebilir veya zayıflatabilir.
3.   **Dolanılık**: İki qubit arasında işlem yapıldığında işlem yapılan qubitin, diğer qubitin durumunu belirleme fenomenidir.

4. **Matrisler**: Kuantum fiziğinde matris işlemleri ile sürekli uğraşılır. O yüzden eğer hatırlatmıyorsanız tekrar etmenizi öneririm.

5. **Karmaşık Sayılar**: Parçacıkların davranışları, sıklıkla karmaşık sayılar kullanılarak matematiksel olarak ifade edilir.!

# **Quantum Kapılar**

Kuantum kapıları, kuantum bilgisayarlar ve kuantum hesaplamalar gibi kuantum mekanik sistemlerde kullanılan mantıksal kapılardır. Kuantum bilgisayarlar, geleneksel bilgisayarların klasik bitlerine benzemeyen kubit adı verilen kuantum bitleri kullandığını belirtmiştik. Kubit'ler, kuantum süperpozisyonu ve kuantum dolanıklık gibi özelliklere sahip olduğu için klasik bilgisayarların işleyişinden önemli ölçüde farklıdır.

**Hadamard Kapısı (H Kapısı)**: Kubit'i klasik bit ve kuantum bit (qubit)
arasında bir süperpozisyon durumuna getirir.

**X Kapısı (Pauli-X Kapısı**): Kubitin durumunu değiştirir, yani 0 durumunda olan bir kubiti 1'e, 1 durumunda olan bir kubiti ise 0'a çevirir. Matematiksel olarak; ***X|0> = |1>, X|1> = |0>*** olur.

**Y Kapısı (Pauli-Y Kapısı)**: Kubit'in durumunu değiştirir ve bir faz kayması ekler. Faz kaymasını tanımladan önce faz değişimini bir tanımlayalım. Faz değişimi dalga fonksiyonlarının değişimidir. Faz kayması ise iki nokta arasındaki faz değişimi farkıdır. Y kapısının matematiksel olarak ifadesi;***Y|0> = i|1>, Y|1> = -i|0>*** olur.

**Z Kapısı (Pauli-Z Kapısı)**: Kubit'in durumunu değiştirir, ancak bir faz faktörü ekler. Faz faktörü eklenmesi ise bir kubitin fazının belirli bir miktarda değiştirilmesini ifade eder. Matematiksel ifadesi;***Z|0> = |0>, Z|1> = -|1>*** olur.

**CNOT Kapısı (Controlled NOT Kapısı)**: İki kubit arasında kontrol edilen bir NOT operasyonu gerçekleştirir. Bir kubit kontrol, diğer kubit hedef olarak kabul edilir ve kontrol kubiti "1" ise, hedef kubiti değiştirir.**kalın metin**

Bir kubitin herhangi bir durumunu 3D uzayda temsil eder ve kubitin temel durumları, kürenin kutuplarına karşılık gelir.
```python
import math
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
qasm_simulator = Aer.get_backend("qasm_simulator")
statevector_simulator = Aer.get_backend("statevector_simulator")
```

```python
def run_on_simulators(circuit):
    statevec_job = execute(circuit, backend=statevector_simulator)
    result = statevec_job.result()
    statevec = result.get_statevector()
    
    num_qubits = circuit.num_qubits
    circuit.measure([i for i in range(num_qubits)], [i for i in range(num_qubits)])
    
    qasm_job = execute(circuit, backend=qasm_simulator, shots=1024).result()
    counts = qasm_job.get_counts()
    
    return statevec, counts
```

```python
circuit = QuantumCircuit(2,2)
statevec, counts = run_on_simulators(circuit)
```

## SuperDense Coding

Superdense coding, kuantum bilgisayarlarındaki bir iletişim protokolüdür. Klasik bilgisayarların aksine, kuantum bilgisayarları kuantum mekanik özelliklerini kullanarak bilgi işleyebilirler. Superdense coding, qubitlerin özel özelliklerini kullanarak bilgi iletimini daha etkili bir şekilde gerçekleştirmeyi amaçlar.

Superdense coding protokolü, Alice ve Bob adlı iki taraf arasındaki bilgi iletimini hızlandırmak için tasarlanmıştır. Temel olarak şu adımları içerir:

**Bağlantı Oluşturma (Entanglement Creation)**: Alice ve Bob'un qubitleri, birbirleriyle dolanık bir duruma getirilir. 
**Alice'in Qubit Hazırlığı:** Alice, iletmek istediği klasik bir bit bilgisini taşıyan bir qubit hazırlar ve kendi qubitini, dolanık durumdaki qubit ile birleştirir.

**Ölçme ve İletim (Measurement and Transmission)**: Alice, kendi qubitini ve dolanık qibiti ölçer. Bu ölçümler, Bob'un qibitinde belirli bir durumun oluşmasını sağlar ve bu durum, Alice'in iletmek istediği klasik bit bilgisini kodlar.

**Bob'un Qubiti Üzerinde Çözme**: Bob, Alice'in qubitini alır ve kendi qubitini kullanarak onun üzerinde ölçüm yapar. Böylece, Alice'in iletmek istediği klasik bilgiyi elde eder.

Superdense coding, klasik iletişim protokollerine kıyasla daha hızlı bilgi iletimine olanak tanıyan bir kuantum iletişim yöntemidir. Bu protokol, kuantum bilgisayarlarının benzersiz özelliklerinden yararlanarak klasik bilgisayarlarla kıyaslandığında daha etkili bir bilgi iletimi sağlamaktadır.

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

def create_superdense_coding_circuit(mesaj):
    qc = QuantumCircuit(2)
    
    # Bell çifti hazırlama
    qc.h(0)
    qc.cx(0, 1)
    
    # Mesajı kodlama
    if mesaj == "00":
        pass  # '00' için hiçbir şey yapma
    elif mesaj == "01":
        qc.x(0)  # '01' için X kapısı uygula
    elif mesaj == "10":
        qc.z(0)  # '10' için Z kapısı uygula
    elif mesaj == "11":
        qc.x(0)  # X kapısı uygula
        qc.z(0)  # Z kapısı uygula
    
    # Bell çifti ölçümü
    qc.cx(0, 1)
    qc.h(0)
    
    # Ölçüm
    qc.measure_all()
    
    return qc

# Gönderilecek mesajı tanımla (2 klasik bit)
message = "11"

# Superdense kodlama devresini oluştur
superdense_coding_circuit = create_superdense_coding_circuit(message)

# Kuantum devresini simüle et
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(superdense_coding_circuit, simulator)
result = simulator.run(assemble(compiled_circuit)).result()

# Ölçüm sonuçlarını görüntüle
counts = result.get_counts()
plot_histogram(counts)
```

## Quantum Teleportation

Quantum teleportation, kuantum bilgisayarları ve kuantum bilgi teorisinde önemli bir konsepttir. Ancak, adından dolayı klasik teleportasyon gibi bir şey değildir. Quantum teleportation, kuantum durumları arasında bilgi transferini ifade eder.

Quantum teleportation protokolü, bir qubit durumunun bir yerden başka bir yere aktarılmasını sağlayan bir kuantum iletişim protokolüdür. Ancak, bu aktarım sırasında klasik bir bilgi transferi gerçekleşir. Yani, kuantum teleportasyonuyla bir qubit'in bir yerden başka bir yere "teleporte" edilmesi, qubit'in kendisi fiziksel olarak taşınmaz; bunun yerine, qubit'in durumu bir yerden diğerine transfer edilir.

Quantum teleportation protokolü şu şekilde çalışır:

**Hazırlık Aşaması (Entanglement)**: İki uzak yerde bulunan partnere ait bir çift entangled (birbirine bağlı) qubit oluşturulur. Bu qubitler birbirleriyle çok güçlü bir korelasyona sahiptir, yani birinde yapılan bir ölçüm, diğerinde de anında bir değişiklik yaratır.

**Gönderici (Alice)**: Alice, teleport etmek istediği qubit ile entangled çiftin bir parçasını birleştirir. Sonrasında bu iki qubiti ölçer ve ölçüm sonucunu klasik bir bilgi olarak Bob'a gönderir.

**Klasik İletişim (Bilgi Transferi)**: Alice'in ölçüm sonucunu içeren klasik bilgi, klasik bir kanal üzerinden Bob'a gönderilir.

**Alıcı (Bob)**: Bob, aldığı klasik bilgiyi kullanarak kendi entangled qubitini manipüle eder. Bu manipülasyon sonucunda, Alice'in orijinal qubitinin durumu Bob'un entangled qubitine transfer edilmiş olur.

Quantum teleportation, kuantum bilgisayarlar arasında bilgi transferi ve kuantum şifreleme protokollerinde önemli bir rol oynar. Ancak, gerçek dünyada uygulanabilirliği için daha gelişmiş kuantum teknolojilerine ihtiyaç vardır.

Peki neden teleportasyona ihtiyaç vardır? **ÇÜNKÜ KUANTUM FİZİĞİ KOPLAYAMAYA İZİN VERMEZ(NO-CLONING THEOREM)**

**No-Cloning Theorem**: 
No-cloning teoremi, kuantum mekaniğinde temel bir kavram olup, herhangi bir belirsiz kuantum durumunun tam bir kopyasını oluşturmanın imkansız olduğunu belirtir. Bu teorem, fizikçi Wootters ve Zurek tarafından 1982'de ilk kez önerilmiştir.

Klasik fizikte, bilgi kopyalamak basit bir süreçtir; bir bilgi parçasını alıp tam bir kopyasını üretebilirsiniz. Ancak, kuantum dünyasında durum, kuantum süperpozisyonu ve ölçüm prensipleri nedeniyle farklıdır.

No-cloning teoremi şu şekilde daha resmi bir ifade edilebilir:

"Herhangi bir bilinmeyen kuantum durumu |ψ⟩ için, |ψ⟩'yi tam olarak bir başka kuantum durumu |φ⟩'ya kopyalayacak bir kuantum operasyonu bulunmamaktadır. Yani, U(|ψ⟩ ⊗ |0⟩) = |ψ⟩ ⊗ |ψ⟩ gibi bir U unitary operatörü yoktur, burada ⊗ tensor çarpımını, |0⟩ ise genellikle yardımcı bir başlangıç durumunu temsil eder."

Burada, ⊗ tensor çarpımını temsil eder ve |0⟩ genellikle yardımcı bir durumu ifade eder. Teorem, belirli bir bilinmeyen kuantum durumunun kesin bir kopyasını oluşturmanın mümkün olmadığını ifade eder.

No-cloning teoremi, kuantum bilgi teorisinin çeşitli alanları için önemli sonuçlara sahiptir, bu alanlara kuantum kriptografi ve kuantum hesaplamalar örnek verilebilir. Örneğin, kuantum anahtar dağıtım protokollerinin temelini, dinleyicilerin algılanmadan kuantum durumlarını kopyalayamayacağı gerçeği oluşturur. Kopyalamanın imkansızlığı, kuantum iletişim sistemlerinin güvenliğini anlamanın önemli bir faktörüdür.

## Bernstein-Vaziranı Algoritması

Gizli bir sayı tuttuğunuzu hayal edin. Klasik bir bilgisayarda bunu ne kadar işlemde tahmin edersiniz? Sayının binary tarzında yazımı kadar. Örneğin elinizde 7 bitlik bir sayı var ve bunu tahmin etmek için 2^7 kere işlem uygulanabilir. Fakat Bernstein-Vazirani algoritması bu sayıyı **tek seferde** tahmin edebilir. Bunun için ***kuantum paralelizasyon*** adı verilen bir yöntem içerir. Kuantum paralelizasyon şöyledir ki; bir kuantum bilgisayarının aynı anda birden çok hesaplama yoluyla potansiyel çözümleri hızlıca değerlendirebilme yeteneğidir. Bu süperpozisyon özelliğinden kaynaklanır. Daha sonra "sihirli kapı" olarak da bilinen bir özel bir oracle kapı işlemi uygulanır. 

Bernstein-Vazirani algoritması, kuantum bilgisayarlarla bir sayıyı hızlı bir şekilde bulma amacını taşıyan bir kuantum algoritmasıdır. Algoritma, klasik bilgisayarlarla bu işlemi gerçekleştirmenin zorluğunu aşmak için tasarlanmıştır.

Algoritmanın İşleyişi:

1. **Sorun Tanımı**: Klasik bir bilgisayar üzerinde bir sayıyı bulmak için genellikle sorgular yaparız. Örneğin, "n" bit uzunluğunda bir gizli sayıyla karşılaştırma yaparak bir sayıyı tespit etmeye çalışırız.

2. **Kuantum Paralelizması**: Bernstein-Vazirani algoritması, kuantum paralelizasyonu kullanarak bu işlemi hızlandırır. Klasik bir bilgisayar, bu sayıyı bulmak için en kötü durumda 2^n sorgu yaparken, kuantum bilgisayar bu işlemi sadece bir sorguda gerçekleştirebilir.

3. **Sorgu Yapısı**: Kuantum sorgusu, bir qubit üzerinde Hadamard dönüşümü uygulanarak yapılır. Ardından, gizli sayıyla ilişkilendirilmiş özel bir oracle (sihirli kapı) kullanılarak sorgu yapılır.

4. **Sonuç Okuma**: Kuantum bilgisayar, klasik bir bilgisayar üzerindeki gibi bir ölçüm yapar. Ancak, kuantum mekaniği prensipleri sayesinde, bu ölçüm, doğru sonucu bulmak için kullanılır.

## Deutsch Algoritması

Deutsch algoritması, kuantum bilgisayarlarının klasik bilgisayarlarla karşılaştırıldığında bazı özel durumlarda daha hızlı çalışabilme yeteneğini göstermek amacıyla tasarlanan bir kuantum algoritmasıdır. Bu algoritma, klasik bir bilgisayarın yaptığı bir işlemi kuantum bilgisayarlarla daha etkili bir şekilde gerçekleştirebilmenin bir örneğidir.

1. **Başlangıç**:
İki qubit kullanılır. İlk qubit başlangıçta ∣0⟩ durumunda bulunurken, ikinci qubit ∣1⟩ durumunda bulunur.\\
\begin{gather}|\psi_0\rangle
=
|0\rangle⊗|1\rangle\end{gather}

2. **Hadamard**:İlk qubit üzerine Hadamard uygulanır  $$|\psi_1\rangle \doteq
 (H⊗I)|\psi_0\rangle$$

3. **Oracle İşlemi**: Oracle, giriş fonksiyonunu temsil eder. Eğer fonksiyon sabitse, ikinci qubiti flip eder; eğer döngüselse herhangi bir işlem yapmaz.
$$|\psi_2\rangle = U_f |\psi_1\rangle$$

4. **Hadamard Dönüşümü ve Ölçüm**: İlk qubit üzerine tekrar Hadamard dönüşümü uygulanır ve ardından ölçüm yapılır.
$$ Ölçüm= |0 \rangle ∧ |1 \rangle $$

Sonuç olarak, eğer ölçüm $$∣0\rangle$$ ise giriş fonksiyonu **sabittir**, $$∣1\rangle$$ ise **döngüseldir**.
Açıklaması şudur; **eğer girdi ile çıktı farklıysa döngüsel, aynı ise sabittir**.
*Deutsch algoritması, klasik bir bilgisayarın bu tür bir problemi çözmesi için en az iki sorgu yapması gereken durumu sadece bir sorgu yaparak çözebilir*. Bu da kuantum bilgisayarlarının belirli durumlarda daha hızlı olabileceğini gösteren bir örnektir.

### Arkasında Yatan Matematik

\begin{gather}
|\psi_1\rangle = \frac{1}{2}(|0\rangle+|1\rangle)⊗(|0\rangle+|1\rangle)\\
=\frac{1}{2}(|00\rangle-|01\rangle+|10\rangle-|11\rangle)\\
\Rightarrow  \frac{1}{2}(|0\rangle|0\oplus f(0)\rangle-|0\rangle|1\oplus f(0)\rangle+|1\rangle+|0\oplus f(1)\rangle-|1\rangle+|1\oplus f(1)\rangle\\
|\psi_2\rangle = \frac{1}{2}(|0\rangle|f(0)\rangle-|0\rangle|1 \oplus f(0)\rangle + |1\rangle +|f(1)\rangle -|1\rangle|1\oplus f(1)\rangle )\\
\\
\Rightarrow EĞER\ F(0)=F(1)\\
\psi_2=\frac{1}{2}(|0\rangle|f(0)\rangle-|0\rangle|1 \oplus f(0)\rangle + |1\rangle +|f(0)\rangle -|1\rangle|1\oplus f(0)\rangle )\\
=\frac{1}{2}((|0\rangle+|1\rangle)|f(0)\rangle)+(-|0\rangle-|1\rangle)|1 \oplus f(1)\rangle)\\
=\frac{1}{2}(|0\rangle+1\rangle)-(|f(0)\rangle-|1\oplus f(1)\rangle)\\
=\frac{1}{\sqrt{2}}(|+\rangle)(|f(0)\rangle-|1 \oplus f(0)\rangle)\\
\psi_3=H(|+\rangle)=|0\rangle\\
\\
\Rightarrow EĞER\ F(0)\neq F(1)\\
|\psi_2 = \frac{1}{2}(|0\rangle|f(0)\rangle-|0\rangle|f(1)\rangle+|1\rangle|f(1)\rangle-|10\rangle|f(0)\rangle)\\
=\frac{1}{2}(|0\rangle⊗|f(0)\rangle-|f(1)\rangle)-|1\rangle⊗(|f(0)\rangle-|f(1)\rangle)\\
=\frac{1}{2}(|0\rangle-|1\rangle)⊗(|f(0)\rangle-|f(1)\rangle)\\
=\frac{1}{\sqrt{2}}(|-\rangle)⊗(|f(0)\rangle-|f(1)\rangle)\\
\psi_3=H(|-\rangle)=|1\rangle\\
\end{gather}

## Shor Algoritması

Shor algoritması özellikle RSA şifrelemesinde kullanılan büyük sayıları faktöriyel etme problemi üzerine odaklanır. Klasik bilgisayarlarla bu tür büyük sayıları faktöriyel etmek, güvenlik sistemlerinde kullanılan birçok şifreleme yöntemini sağlam hale getirir. Ancak Shor algoritması, bu tür problemleri kuantum bilgisayarlarla çok daha hızlı bir şekilde çözebilir.

Shor algoritması, büyük sayıları çarpanlarına ayırmak için kuantum süperpozisyon ve kuantum entanglement özelliklerini kullanır. Bu sayede, klasik algoritmalarla günler veya aylar sürebilecek işlemleri, kuantum bilgisayarlarla çok daha hızlı bir şekilde gerçekleştirebilir.

1. **Faktöriyel Edilecek Sayı Seçim**: İlk olarak, faktöriyel etmek istediğiniz büyük sayıyı seçersiniz. Bu sayı genellikle iki büyük asal çarpana ayrılmış bir sayıdır (örneğin, RSA şifreleme sisteminde kullanılan modüller).
2. **Kuantum Devresi Hazırlığı** :İlk kuantum devresi, ikinci bir kuantum devresini çağıracak bir kontrol devresini içerir. Bu kontrol devresi, modüler üs alma işlemlerini yönetir.
3. **İkinci Kuantum Devresi**: İkinci kuantum devresi, modüler üs alma işlemlerini gerçekleştirir ve bu işlemler klasik bilgisayarlarla zor veya zaman alıcı olan adımlardır.
4. **QFT (Quantum Fourier Transform) Uygulaması**:Kuantum Fourier Dönüşümü, kuantum durumlarını Fourier dönüşümüne tabi tutan bir operatördür. Bu adım, özellikle büyük sayıların periyodik fonksiyonlarını analiz etmeye yardımcı olur.
5. **Periyodun Bulunması**: QFT uygulaması sonucunda elde edilen kuantum durumları, sayının periyodik bir fonksiyon olmasına yol açar. Bu periyodu bulmak, faktörleri elde etmede kritik bir adımdır.
6. **Klasik Hesaplama İle Faktör Bulma**:Periyodu elde ettikten sonra, klasik bir bilgisayar kullanılarak bu periyodu kullanarak sayının faktörleri hesaplanır.

Shor algoritması, bu adımların bir kombinasyonu ile çalışır ve özellikle büyük asal sayılardan türetilmiş modüler aritmetik işlemleri üzerinde kuantum bilgisayarlarının avantajlarını kullanır. Bu algoritma, klasik bilgisayarlarla çözülmesi çok zor olan problemleri çözmek için tasarlanmış olup, bu tür problemlerin çözüm süresini dramatik bir şekilde azaltabilir.

### QFT(KUANTUM FOURIER DÖNÜŞÜMÜ)

Bu işlem, hesaplama bazından Fourier bazına geçiş demektir. Hesaplama bazı; Q-Sphere'de yukarıda 0, aşağıda 1 gördüğümüz bazdır. Fourier bazı ise |+⟩ ve |-⟩ şeklinde gördüğümüz bazlardır.(|0⟩ için |+⟩,|1⟩ ve |-⟩). **fourier bazında olmak illa süperpozisyonda veya tam olarak + ve - olmasına da gelmez**. Bunun nedeni farklı açılarda olabileceği ihtimalidir. Bunu formülize edecek olursak;

\begin{gather}
QFT|x\rangle= |\widetilde{x}\rangle=\frac{1}{\sqrt{N}}\sum^{N-1}_{y=0}e^{\frac{2\pi i xy}{N}}|y \rangle\\NOT:\
N = 2^n, \ n = Qubit, \ e^{\pi*i} = -1
\end{gather}

1 qubit için uygulayalım.
\begin{gather}
N = 2^1 =2\\
\frac{1}{\sqrt{2}}\sum^{2-1}_{y=0}e^{\frac{2\pi xy}{2}}|y \rangle \\
=\frac{1}{\sqrt{2}}e^{\frac{2\pi i x 0}{2}}|0\rangle + \frac{1}{\sqrt{2}}e^{\frac{2\pi i x 1}{2}}|1\rangle \\
=\frac{1}{\sqrt{2}}|0\rangle  + \frac{1}{\sqrt{2}}e^{\pi i x }|1\rangle \\
Eğer \ x = 0 \rightarrow \frac{1}{\sqrt{2}}|0\rangle + |1\rangle = |+\rangle\\
Eğer \ x = 1 \rightarrow \frac{1}{\sqrt{2}}|0\rangle - |1\rangle = |-\rangle\\
\end{gather}
